## **Face Detectors**

```python
backends = [
  'opencv',
  'ssd',
  'dlib',
  'mtcnn',
  'retinaface',
  'mediapipe',
  'yolov8',
  'yunet',
]

#face verification
fv = DeepFace.verify(img1_path = "/content/jp.jpg",
        img2_path = "/content/kimk.jpg",
        detector_backend = backends[0]
)

#face recognition
fr = DeepFace.find(img_path = "/content/james.png",
        db_path = "/content",
        detector_backend = backends[1]
)

#embeddings
embedding_objs = DeepFace.represent(img_path = "/content/rambo.jpg",
        detector_backend = backends[2]
)

#facial analysis
fa = DeepFace.analyze(img_path = "/content/download.jpg",
        detector_backend = backends[3]
)

#face detection and alignment
face_objs = DeepFace.extract_faces(img_path = "/content/rambo.jpg",
        target_size = (224, 224),
        detector_backend = backends[4]
)
```
